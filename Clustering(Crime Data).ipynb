```python
# Normalization function # use sklearn library norm
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return x
```

```python
# Normalized data frame (considering the numerical part of data)
df_norm = norm_func(crime.iloc[:,1:])
df_norm
```

```python
# create dendrogram
dendrogram = sch.dendrogram(sch.linkage(df_norm, method='complete'))
```

```python
dendrogram = sch.dendrogram(sch.linkage(df_norm, method='average'))
```

```python
dendrogram = sch.dendrogram(sch.linkage(df_norm, method='single'))
```

```python
dendrogram = sch.dendrogram(sch.linkage(df_norm, method='centroid'))
```

```python
# create clusters
hc = AgglomerativeClustering(n_clusters=3, affinity = 'euclidean', linkage = 'complete')
```

```python
hc.fit(df_norm)
```

```python
# save clusters for chart
y_hc = hc.fit_predict(df_norm)
y_hc
```

```python
clusters=pd.DataFrame(y_hc,columns=['Clusters'])
clusters
```

```python
crime['h_clusterid'] = y_hc
crime
```

**K-Means Clustering**

```python
from sklearn.cluster import KMeans
```

```python
crime1=pd.read_csv('crime_data.csv')
crime1.head()
```

```python
# Standardisation function 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_crime1_df = scaler.fit_transform(crime1.iloc[:,1:])
```

```python
scaled_crime1_df
```

```python
#The K-means algorithm aims to choose centroids that minimise the inertia, or within-cluster sum-of-squares criterion:
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(scaled_crime1_df)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# selecting 4 clusters from above scree plot
model=KMeans(n_clusters=4)
model.fit(scaled_crime1_df)
model.labels_
```

```python
x=pd.Series(model.labels_)
crime1['Cluster']=x
```

```python
crime1
```

```python
crime1.iloc[:,1:5].groupby(crime1.Cluster).mean()
```

**DBSCAN**

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
crime2=pd.read_csv('crime_data.csv')
crime2.head()
```

```python
crime2.info()
```

```python
df=crime2.iloc[:,1:5]
df.values
```

```python
stscaler=StandardScaler().fit(df.values)
x=stscaler.transform(df.values)
x
```

```python
dbscan=DBSCAN(eps=2,min_samples=5)
dbscan.fit(x)
```

```python
dbscan.labels_
```

```python
#Noisy samples are given the label -1.
cl=pd.DataFrame(dbscan.labels_,columns=['cluster'])
cl
```

```python
pd.concat([crime2,cl],axis=1)
```
